```python
#Import Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Extract CSVs into DataFrames
csv_file = "C:/Users/bdsan/Desktop/ETL project files/Resources/Orlando_Crimes.csv"
orl_data = pd.read_csv(csv_file)
orl_data.head()
```

```python
# Drop extraneous columns
orl_data.drop(['Case Number', 'Case Disposition', 'Case Offense Type', 'Status', 'Location'], axis=1, inplace=True)
```

```python
orl_data.head()
```

```python
#View data types
orl_data.dtypes
```

```python
#Rename columns 
orl_data.rename(columns={'Case Date Time': 'Date', 'Case Location': 'Location', 'Case Offense Location Type': 'Premise','Case Offense Category': 'Description'}, inplace=True)
```

```python
orl_data.head()
```

```python
#Add new column for City
orl_data["City"] = "Orlando"
orl_data.head()
```

```python
#Add new column for State
orl_data["State"] = "Florida"
orl_data.head()
```

```python
# Reorganizing the columns using double brackets
organized_df = orl_data[["Date","Location","City","State", "Description", "Premise"]]
organized_df.head()
```

```python
#Remove the time
organized_df['Date'] = organized_df['Date'].str.split(' ',expand=True)[0]
```

```python
organized_df.head()
```
