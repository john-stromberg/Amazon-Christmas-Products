```python
#Import Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Extract CSVs into DataFrames
csv_file =r"C:\Users\johns\Desktop\Crime-Data-ETL-Project\Resources\Orlando_Crimes_Raw.csv"
orl_data = pd.read_csv(csv_file)
orl_data.head()
```

```python
# Drop extraneous columns
orl_data.drop(['Case Number', 'Case Disposition', 'Case Offense Type', 'Status', 'Location'], axis=1, inplace=True)
```

```python
orl_data.head()
```

```python
#View data types
orl_data.dtypes
```

```python
#Rename columns 
orl_data.rename(columns={'Case Date Time': 'crime_date', 'Case Location': 'location', 'Case Offense Location Type': 'premise','Case Offense Category': 'description'}, inplace=True)
```

```python
orl_data.head()
```

```python
#Add new column for City
orl_data["city"] = "Orlando"
orl_data.head()
```

```python
#Add new column for State
orl_data["state"] = "Florida"
orl_data.head()
```

```python
# Reorganizing the columns using double brackets
organized_df = orl_data[["crime_date","location","city","state", "description", "premise"]]
organized_df.head()
```

```python
#Remove the time
organized_df['crime_date'] = organized_df['crime_date'].str.split(' ',expand=True)[0]
```

```python
#convert columndate to date/time data type
organized_df['crime_date'] = pd.to_datetime(organized_df['crime_date'])
```

```python
organized_df.head()
```

```python
connection_string = "postgres:2231jack.@localhost:5432/Crime_Data"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
organized_df.to_sql(name="orlando", con=engine, if_exists="append", index= False)
```
